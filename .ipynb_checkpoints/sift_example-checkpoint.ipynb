## Bag of Visual Words Example

```python
import cv2
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
%matplotlib inline
```

```python
PATH = 'data/dog-breed/test/'
```

```python
nfeatures = 0
extractor = cv2.xfeatures2d.SIFT_create(nfeatures=nfeatures)
```

```python
def kpt_image(image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    keypoints, descriptors = extractor.detectAndCompute(gray_image, None)
    return cv2.drawKeypoints(image, keypoints, None)    
```

```python
for p in list(Path(PATH).iterdir())[: 5]:
    image = cv2.imread(str(p))
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    fig, axes = plt.subplots(1, 2, figsize=(8, 5))
    axes[0].imshow(image)
    axes[1].imshow(kpt_image(image))
    plt.show()
```
